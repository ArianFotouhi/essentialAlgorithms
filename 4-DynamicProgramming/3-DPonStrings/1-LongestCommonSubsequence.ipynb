# 1143 - Longest Common Subsequence

Given two strings text1 and text2, return the length of their longest common subsequence. If there is no common subsequence, return 0.

A subsequence of a string is a new string generated from the original string with some characters (can be none) deleted without changing the relative order of the remaining characters.

For example, "ace" is a subsequence of "abcde".
A common subsequence of two strings is a subsequence that is common to both strings.

## Example 1:

- Input: text1 = "abcde", text2 = "ace" 
- Output: 3  
- Explanation: The longest common subsequence is "ace" and its length is 3.

## Example 2:

- Input: text1 = "abc", text2 = "abc"
- Output: 3
- Explanation: The longest common subsequence is "abc" and its length is 3.

## Example 3:

- Input: text1 = "abc", text2 = "def"
- Output: 0
- Explanation: There is no such common subsequence, so the result is 0.

# Base Version

```python
def longestCommonSubsequence(text1: str, text2: str) -> int:
    n, m = len(text1), len(text2)
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if text1[i - 1] == text2[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[n][m]
longestCommonSubsequence("abcde", "ace")  # 3 ("ace")
```

# Verbose Version

```python
def lcs_verbose(text1: str, text2: str) -> int:
    """
    Returns the length of the LCS between text1 and text2,
    while VERBOSE-printing each DP transition and the table state.
    """
    n, m = len(text1), len(text2)

    # dp[i][j] = LCS length for text1[:i] and text2[:j]
    dp = [[0]*(m+1) for _ in range(n+1)]

    def print_table(dp, i=None, j=None):
        """Pretty-print the DP table with row/col headers.
        Highlights the most recently updated cell (i, j) with brackets.
        """
        # Column header
        header = ["    "] + ["  ''"] + [f"  {c}" for c in text2]
        print("".join(header))
        for r in range(n+1):
            row_label = " ''" if r == 0 else f" {text1[r-1]}"
            line = [f"{row_label:>4}"]
            for c in range(m+1):
                val = dp[r][c]
                if i == r and j == c:
                    cell = f"[{val}]"
                else:
                    cell = f" {val} "
                line.append(f"{cell:>4}")
            print("".join(line))
        print()

    # Initial table
    print("Initial DP table (all zeros):")
    print_table(dp)

    # Fill dp
    for i in range(1, n+1):
        for j in range(1, m+1):
            a = text1[i-1]
            b = text2[j-1]
            if a == b:
                dp[i][j] = 1 + dp[i-1][j-1]
                print(f"i={i}, j={j} | text1[i-1]='{a}' == text2[j-1]='{b}' → match")
                print(f"dp[{i}][{j}] = 1 + dp[{i-1}][{j-1}] = 1 + {dp[i-1][j-1]} = {dp[i][j]}")
            else:
                up = dp[i-1][j]
                left = dp[i][j-1]
                dp[i][j] = max(up, left)
                choice = "up" if up >= left else "left"
                print(f"i={i}, j={j} | '{a}' != '{b}' → no match")
                print(f"dp[{i}][{j}] = max(dp[{i-1}][{j}]={up}, dp[{i}][{j-1}]={left}) = {dp[i][j]} (from {choice})")
            print_table(dp, i, j)

    # Optional: reconstruct one LCS for illustration
    lcs_len = dp[n][m]
    print(f"LCS length = {lcs_len}")
    print("Reconstructing one LCS (optional):")
    i, j = n, m
    chars = []
    while i > 0 and j > 0:
        if text1[i-1] == text2[j-1]:
            chars.append(text1[i-1])
            print(f"Take '{text1[i-1]}' from i={i}, j={j} (diag).")
            i -= 1
            j -= 1
        else:
            if dp[i-1][j] >= dp[i][j-1]:
                print(f"Move up from (i={i}, j={j}) to (i={i-1}, j={j}).")
                i -= 1
            else:
                print(f"Move left from (i={i}, j={j}) to (i={i}, j={j-1}).")
                j -= 1
    lcs_str = "".join(reversed(chars))
    print(f"One LCS: '{lcs_str}'")
    return lcs_len

# ---------- Small demo ----------
if __name__ == "__main__":
    # Keep these short to make the verbose output readable.
    a = "abcde"
    b = "ace"
    print(f"text1 = '{a}', text2 = '{b}'\n")
    ans = lcs_verbose(a, b)
    print(f"\nAnswer: {ans}")
```
